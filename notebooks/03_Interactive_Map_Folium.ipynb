## Let's mark the success/failed launches for each site on the map


Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.



```python
spacex_df.tail(10)
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

```python
spacex_df['marker_color'] = ['green' if x == 1 else 'red' for x in spacex_df['class']]
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
```

```python
# Add the marker cluster to the site map to group launch markers together
site_map.add_child(marker_cluster)

# Iterate through each launch record in the SpaceX dataframe to create markers
for index, record in spacex_df.iterrows():
    # Define the coordinate for the specific launch using its Latitude and Longitude
    coordinate = [record['Lat'], record['Long']]
    
    # Create a marker for the launch with a specific color based on the launch outcome
    marker = folium.Marker(
        coordinate,
        icon = folium.Icon(
            color='white', 
            icon_color=record['marker_color']
        )
    )
    
    # Add the newly created marker to the marker cluster
    marker_cluster.add_child(marker)

# Display the updated site map
site_map
```

From the color-labeled markers in marker clusters, we can easily identify which launch sites have relatively high success rates.


## Let's now calculate the distances between a launch site to its proximities


We need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while we are exploring the map, we can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

This function implements the Haversine formula, which is the standard way to calculate the great-circle distance between two points on a sphere given their longitudes and latitudes.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
launch_site_lat = 34.632834	
launch_site_lon = -120.610745
coastline_lat = 34.6339
coastline_lon = - 120.626
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
```

```python
# Define the coordinate for the specific point of interest on the coastline
location = [coastline_lat, coastline_lon]

# Create a marker at the coastline coordinate with a text label showing the calculated distance in kilometers
distance_marker = folium.Marker(
    location,
    icon=DivIcon(
        # Set the size and alignment of the text label
        icon_size = (20,20),
        icon_anchor=(0,0),
        # Format the distance value to two decimal places and apply styling to the HTML label
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
    )
)
```

```python
locations = [[launch_site_lat, launch_site_lon],[coastline_lat, coastline_lon]]
lines = folium.PolyLine(locations, weight =1)
site_map.add_child(lines)
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
railway_lat = 34.63334
railway_lon = -120.6396
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)

location = [railway_lat, railway_lon]
distance_marker = folium.Marker(
    location = location,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_railway),
    )
)

locations = [[launch_site_lat, launch_site_lon], [railway_lat, railway_lon]]
lines = folium.PolyLine(locations, weight = 1)
site_map.add_child(lines)
```
